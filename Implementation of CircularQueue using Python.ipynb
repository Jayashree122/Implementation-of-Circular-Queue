```python
#Implementation of CircularQueue
class CircularQueue:
    def __init__(self,size):
        self.q = [None]*size
        self.f = self.r = -1
        self.size = size
    def isempty(self):
        #Time Complexity: O(1)
        return self.f == -1 and self.r == -1
    def isfull(self):
        #Time Complexity: O(1)
        return (self.r+1)%self.size == self.f
    def enqueue(self,ele):
        #Time Complexity: O(1)
        if self.isfull():
            raise Exception("Queue is full")
        elif  self.isempty():
            self.f = self.r = 0
            self.q[self.r] = ele
        else:
            self.r = (self.r+1)%self.size
            self.q[self.r] = ele
    def dequeue(self):
        #Time Complexity: O(1)
        if self.isempty():
            raise Exception("Queue is empty")
        elif self.f == self.r:
            t = self.q[self.f]
            self.f = self.r = -1
            return t
        else:
            t = self.q[self.f]
            self.f = (self.f+1)%self.size
            return t
    def len(self):
        #Time Complexity: O(1)
        h = self.r if self.r>self.f else self.f
        return abs(abs(h - self.f) - abs(h -self.r))+1;
    def traverse(self):
        #Time Complexity: O(n)
        if self.isempty():
            raise Exception("Queue is empty")
        i = self.f
        while i!=self.r:
            print(self.q[i],end = " ")
            i = (i+1)%self.size
        print(self.q[i],end = " ")
#Space Complexity: O(n)
```

```python
b=CircularQueue(5)
b.enqueue(10)
b.enqueue(20)
b.enqueue(30)
b.enqueue(40)
b.enqueue(50)
b.dequeue()
b.enqueue(60)
b.dequeue()
b.traverse()
```
